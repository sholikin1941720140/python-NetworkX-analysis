```python
import networkx as nx
import matplotlib.pyplot as plt

# Data dari studi kasus
data = [
    {"Nama": "Bellavania Roufin", "Divisi": "Divisi Acara sebagai MC", "Interaksi": ["Regent", "Monic", "Fereyn", "Andrew", "Toby", "Nadine", "Nadhif"]},
    {"Nama": "Nadine", "Divisi": "Divisi Acara", "Interaksi": ["Joe Shane"]},
    {"Nama": "Regentino Rizky Gunawan", "Divisi": "Ketua panitia", "Interaksi": ["Bella - Acara", "Toby - Acara", "Andrew - Acara", "Rozy - Pubdok", "Cisca - Bendahara", "Anggi - Sekretaris"]},
    {"Nama": "Francisca", "Divisi": "Bendahara", "Interaksi": []},
    {"Nama": "Abdul Kadir Zaelani", "Divisi": "Konsumsi", "Interaksi": ["Regent"]},
    {"Nama": "Fereyn Al ikhsan", "Divisi": "Panitia seksi acara", "Interaksi": ["Bella", "Regen", "Monic", "Andrew", "Lutfi", "Sisca", "Toby"]},
    {"Nama": "Rozi", "Divisi": "Media", "Interaksi": ["Regen", "Monica", "Anggi", "Toby", "Aul", "Nadine", "Fereyn", "Bella", "Andrew", "Lutfhy", "Kadir", "Sisca", "Evi", "Ara", "Khalse", "Ardi"]},
    {"Nama": "Evi", "Divisi": "Administrasi/Bendahara", "Interaksi": ["Regen", "Monica", "Rozi", "Budi", "Joe"]},
    {"Nama": "Asmad", "Divisi": "Seksi Acara", "Interaksi": ["Aisyah", "Joe", "Kevin"]},
    {"Nama": "Roy gantang", "Divisi": "Acara", "Interaksi": ["Aisyah", "Joe", "Risa", "Rizki"]},
    {"Nama": "Andrew", "Divisi": "Seksi Acara", "Interaksi": ["Regen", "Monica", "Anggi", "Toby", "Aul", "Nadine", "Fereyn", "Bella", "Andrew", "Lutfhy", "Kadir", "Sisca", "Evi", "Ara", "Khalse", "Ardi", "Joe", "Aisyah", "Pemilik Makato"]}
]

# Membuat graf dengan NetworkX
G = nx.Graph()

# Menambahkan nodes dan edges
for person in data:
    G.add_node(person["Nama"])
    for interaction in person["Interaksi"]:
        G.add_edge(person["Nama"], interaction)

# Visualisasi graf
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_size=8, node_size=700, node_color='skyblue', font_color='black', font_weight='bold')
plt.title("Graf Interaksi Panitia Acara 'Tiba-tiba 1th Anniversary'")
plt.show()

# Menghitung dan mencetak sentralitas
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)

print("\nDegree Centrality:")
for person, centrality in degree_centrality.items():
    print(f"{person}: {centrality}")

print("\nCloseness Centrality:")
for person, centrality in closeness_centrality.items():
    print(f"{person}: {centrality}")

print("\nBetweenness Centrality:")
for person, centrality in betweenness_centrality.items():
    print(f"{person}: {centrality}")

print("\nEigenvector Centrality:")
for person, centrality in eigenvector_centrality.items():
    print(f"{person}: {centrality}")
```
