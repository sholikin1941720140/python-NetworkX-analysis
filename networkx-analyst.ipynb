```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community

# Membuat jaringan sosial (graph) dengan data dummy
G = nx.Graph()

# Menambahkan nama-nama acak sebagai simpul
nama_user = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Harry", "Ivy", "Jack",
             "Katherine", "Leo", "Mia", "Nathan", "Olivia"]

G.add_nodes_from(nama_user)

# Menambahkan edge (hubungan) antara nama user secara acak
for i in range(len(nama_user)):
    for j in range(i + 1, len(nama_user)):
        if i != j and not G.has_edge(nama_user[i], nama_user[j]):
            G.add_edge(nama_user[i], nama_user[j])

# Memastikan jaringan terhubung sepenuhnya
if not nx.is_connected(G):
    # Menggunakan largest_connected_nodes untuk mempertahankan subgraf terbesar yang terhubung
    largest_connected_nodes = max(nx.connected_components(G), key=len)
    G = G.subgraph(largest_connected_nodes).copy()

# Menghitung beberapa metrik jaringan
diameter = nx.diameter(G)
density = nx.density(G)
reciprocity = nx.reciprocity(G)

# Menghitung modularity
partition = community.greedy_modularity_communities(G)
modularity = community.modularity(G, partition)

# Menghitung metrik level aktor
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)

# Menampilkan hasil metrik jaringan
print("Metrik Jaringan:")
print("Diameter: {}".format(diameter))
print("Density: {}".format(density))
print("Reciprocity: {}".format(reciprocity))
print("Modularity: {}".format(modularity))

# Menampilkan hasil metrik level aktor
print("\nMetrik Level Aktor:")
print("Degree Centrality:")
for node, value in degree_centrality.items():
    print(f"{node}: {value}")

print("\nCloseness Centrality:")
for node, value in closeness_centrality.items():
    print(f"{node}: {value}")

print("\nBetweenness Centrality:")
for node, value in betweenness_centrality.items():
    print(f"{node}: {value}")

print("\nEigenvector Centrality:")
for node, value in eigenvector_centrality.items():
    print(f"{node}: {value}")

# Visualisasi Grafik
pos = nx.spring_layout(G)  # Menentukan posisi simpul untuk visualisasi
nx.draw(G, pos, with_labels=True, font_size=8, node_size=500, node_color='skyblue', font_color='black', font_weight='bold', edge_color='gray')
plt.title("Jaringan Sosial")
plt.show()
```
